```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/Users/jialechen/Desktop/PhD/CT/Pang_2022_GenomeBiol_3D/results/CT parameters/All_Top_parameters.csv')

df = pd.DataFrame(data)
df.head()
```

```python
# 设置样式
sns.set(style="whitegrid")

# 绘制平行、系列和交叉百分比的堆叠条形图
plt.figure(figsize=(10, 6))
sns.barplot(x="chromosome", y="Parallel (%)", data=df, color="blue", label="Parallel")
sns.barplot(x="chromosome", y="Series (%)", data=df, color="green", bottom=df["Parallel (%)"], label="Series")
sns.barplot(x="chromosome", y="Cross (%)", data=df, color="red", bottom=df["Parallel (%)"] + df["Series (%)"], label="Cross")
plt.title("Parallel, Series, and Cross Percentages for Each Chromosome")
plt.xlabel("Chromosome")
plt.ylabel("Percentage")
plt.legend()
plt.show()
```

```python
# 绘制染色体之间联系数的柱状图
plt.figure(figsize=(10, 6))
sns.barplot(x="chromosome", y="N contacts", data=df, color="purple")
plt.title("Number of Contacts for Each Chromosome")
plt.xlabel("Chromosome")
plt.ylabel("Number of Contacts")
plt.show()
```

```python
# 绘制分形维度的散点图
plt.figure(figsize=(10, 6))
sns.scatterplot(x="chromosome", y="Fractal dimension", data=df, color="orange")
plt.title("Fractal Dimension for Each Chromosome")
plt.xlabel("Chromosome")
plt.ylabel("Fractal Dimension")
plt.show()
```

```python
# 绘制r平方值的折线图
plt.figure(figsize=(10, 6))
sns.lineplot(x="chromosome", y="r squared", data=df, marker="o", color="brown")
plt.title("r Squared Values for Each Chromosome")
plt.xlabel("Chromosome")
plt.ylabel("r Squared")
plt.show()
```

```python
# 绘制平行、系列和交叉百分比的堆叠条形图，使用FacetGrid分别绘制每个细胞的数据
g = sns.FacetGrid(df, col="Sample", height=6, margin_titles=True)
g.map_dataframe(sns.barplot, x="chromosome", y="Parallel (%)", color="blue", label="Parallel")
g.map_dataframe(sns.barplot, x="chromosome", y="Series (%)", color="green", bottom=df["Parallel (%)"], label="Series")
g.map_dataframe(sns.barplot, x="chromosome", y="Cross (%)", color="red", bottom=df["Parallel (%)"] + df["Series (%)"], label="Cross")
g.set_axis_labels("chromosome", "Percentage")
g.add_legend(title="Contact Type")
plt.suptitle("Parallel, Series, and Cross Percentages for Each Chromosome", size=16)
plt.show()
```
